```python
import os
import pandas as pd
import json


result = pd.DataFrame(columns=[
    'file_name', 'trend_def', 'features', 'model_type', 'predict_steps', 'weight_after_reversal',  'trend_accuracy', 
    'trend_precision', 'trend_recall', 'trend_f1_score', 'reversal_accuracy', 'reversal_precision', 
    'reversal_recall', 'reversal_f1_score', 'signal_accuracy', 'signal_precision', 'signal_recall', 'signal_f1_score', 
    'roc_auc', 'pred_days_difference_abs_mean', 'pred_in_advance', 'execution_time', 'learning_rate'])

root_path = 'DNN_Projects_l2_xavier_learningRate_epoch'
for folder in os.listdir(root_path):
    first_path = os.path.join(root_path, folder)
    for subfolder in os.listdir(first_path):
        second_path = os.path.join(first_path, subfolder)
        print(second_path)
        summary_path = os.path.join(second_path, 'reports/summary.json')
        if not os.path.exists(summary_path):
            continue
        with open(summary_path, 'r') as file:
            trend_def = folder.split('-')[0]
            features = subfolder.split('-')[0]
            content = json.loads(file.read())
            trend_confusion_matrix_info = json.loads(content['trend_confusion_matrix_info'])
            reversed_trend_confusion_matrix_info = json.loads(content['reversed_trend_confusion_matrix_info'])
            signal_confusion_matrix_info = json.loads(content['signal_confusion_matrix_info'])
            
            # Create a DataFrame for the new row
            newrow = pd.DataFrame({
                'file_name': [folder],
                'trend_def': [trend_def],
                'features': [features],
                'model_type': [content['usingData']['model_type']],
                'predict_steps': [content['usingData']['predict_steps']],
                'weight_after_reversal': [content['usingData']['weight_after_reversal']],
                'trend_accuracy': [trend_confusion_matrix_info['Accuracy']['0']],
                'trend_precision': [trend_confusion_matrix_info['Precision']['0']],
                'trend_recall': [trend_confusion_matrix_info['Recall']['0']],
                'trend_f1_score': [trend_confusion_matrix_info['F1 Score']['0']],
                'reversal_accuracy': [reversed_trend_confusion_matrix_info['Accuracy']['0']],
                'reversal_precision': [reversed_trend_confusion_matrix_info['Precision']['0']],
                'reversal_recall': [reversed_trend_confusion_matrix_info['Recall']['0']],
                'reversal_f1_score': [reversed_trend_confusion_matrix_info['F1 Score']['0']],
                'signal_accuracy': [signal_confusion_matrix_info['Accuracy']['0']],
                'signal_precision': [signal_confusion_matrix_info['Precision']['0']],
                'signal_recall': [signal_confusion_matrix_info['Recall']['0']],
                'signal_f1_score': [signal_confusion_matrix_info['F1 Score']['0']],
                'roc_auc': [float(content['roc_auc'])],
                'pred_days_difference_abs_mean': [float(content['pred_days_difference_abs_mean'])],
                'pred_in_advance': [int(content['pred_in_advance'])],
                'execution_time': [float(content['execution_time'])],
                'learning_rate': [content['usingData']['learning_rate']]
            })

            # Concatenate the new row with 'result'
            result = pd.concat([result, newrow], ignore_index=True)  # Ensure the index is reset
            
```

```python
result.sort_values(by=['reversal_f1_score'], ascending=False, inplace=True)
result['reversal_f1_score']
```

```python
result['name'] = result['trend_def']+',\nlearning_rate: '+ result['learning_rate'].astype(str)
```

```python
import matplotlib.pyplot as plt
result.plot(x='name', y=['reversal_f1_score', 'trend_f1_score'], kind='bar', figsize=(20, 6))

# Annotate the y-axis values on the bars for "F1 Score"
for index, value in enumerate(result['reversal_f1_score']):
    plt.annotate(str(round(value, 2)), xy=(index, value), ha='center', va='bottom')

# Annotate the y-axis values on the bars for "Total return"
for index, value in enumerate(result['trend_f1_score']):
    plt.annotate(str(round(value, 2)), xy=(index, value), ha='center', va='bottom')

plt.title('Reversal F1 Score for Different Trend Definitions and Weight After Reversal')
plt.xticks(rotation=90)  # Rotate x-axis labels to be horizontal
plt.show()

```
